```python
#pri_toys
features = []; cat_features = []; not_features = []; nums=[]
for k in Train.columns[1:]:
    if Train[k].isnull().sum() < 6000:
        features.append(k)
        if Train[k].dtype == 'O':
            cat_features.append(k)
            print('There is '+ str(len(Train[k].value_counts()))+' Class in: ' +k)
        else:
          nums.append(k)
    else:
        not_features.append(k)

print('----------------------------------')
print('We have '+str(len(features)) + ' features')
print('We have '+str(len(cat_features)) + ' categorical features')
print('We have '+str(len(not_features)) + ' features that have more than 6000 of missing values')
```

```python
y = train.target

num_cols = list(set(test.select_dtypes('number')).intersection(df.select_dtypes('number')))
X = df[num_cols]
X = X.fillna(0)

# Split data into training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1234)

# Train a model
#model = LinearRegression()
model = XGBRegressor(n_estimators = 1000,random_state=42)

model.fit(X_train, y_train,eval_set=[(X_test,y_test)],verbose = 20,early_stopping_rounds=300)

# Make predictions
y_pred = model.predict(X_test)

# Check score
mean_squared_error(y_test, y_pred, squared=False)
```
